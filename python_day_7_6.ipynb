```python
def square(n):
     for i in range(n):
            yield i**2
```

```python
out=square(5)
```

```python
%%writefile mathfile.py


def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def multiply(a,b):
    return a*b
```

```python
%%writefile test_mathfile.py

import mathfile

def test_add():
    output=mathfile.add(2,3)
    assert output==5
    
def test_sub():
    output=mathfile.sub(4,3)
    assert output==1

def test_multiply():
    output=mathfile.multiply(2,3)
    assert output==6
```

```python
%%writefile rect_calc.py

def area(a,b):
    return a*b
def perimeter(a,b):
    return 2*a+2*b
```

```python
%%writefile test_rect_calc.py

import rect_calc

def test_area():
    output=rect_calc.area(5,10)
    assert output==50
def test_perimeter():
    output=rect_calc.perimeter(4,10)
    assert output==28
```
